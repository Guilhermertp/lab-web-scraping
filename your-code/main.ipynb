#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'

html = requests.get(url).text

soup = BeautifulSoup(html, "lxml")
````

````python
soup=soup.find_all('td', attrs={'class':'titleColumn'})
````

````python

links = [element.find_all('a') for element in soup]
links
````

````python
movies = [element.text for link in links for element in link]
movies
````

````python
date = [element.find_all('span', attrs ={'class':'secondaryInfo'}) for element in soup]

date = [element.text.replace('(','').replace(')','') for link in date for element in link]
date
````

````python
director_stars = [item.get('title') for link in links for item in link]
#director_stars
````

````python
lst = pd.DataFrame(list(zip(movies,date,director_stars)),columns =['movies', 'year', 'director and stars']) 
lst
````
